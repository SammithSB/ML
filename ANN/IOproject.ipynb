```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing data
# We import the csv file
data = pd.read_csv('data-2.csv')
del data['Unnamed: 32']
```

```python
# We choose the columns we use as parameters to predict the value
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values

# Encoding categorical data
# We encode the data as we need to pass integer values 
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0)

#Feature Scaling
# It is done to make sure no variables are dominative over other and values are within a range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# importing necessary libraries and functions from them
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
# Initialising the ANN
# We use sequential for initialising the neural network
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
# We need to mention the input dimension for the first layer as the neural net should know what dimension it should expext
classifier.add(Dense(output_dim=16, init='uniform', activation='relu', input_dim=30))
# Adding dropout to prevent overfitting
classifier.add(Dropout(p=0.1))
```

```python
# Adding the second hidden layer
classifier.add(Dense(output_dim=16, init='uniform', activation='relu'))
# Adding dropout to prevent overfitting
classifier.add(Dropout(p=0.1))
```

```python
# Adding the output layer
# We want the probability of the value being 0 or 1 so we choose sigmoid function
classifier.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))
```

```python
# Compiling the ANN
# We choose adam optimizer and bunary_crossentropy as loss function
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Fitting the ANN to the Training set
#We choose batch size and number of epochs 
classifier.fit(X_train, y_train, batch_size=100, nb_epoch=200)
```

```python
# Predicting the Test set results. We make condition to be >0.5 as for value to be 1, it should be >0.5
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0]))*100))
```

```python
#Adding a heat map to differentiate true values and false values
sns.heatmap(cm,annot=True)
plt.savefig('h.png')
```
